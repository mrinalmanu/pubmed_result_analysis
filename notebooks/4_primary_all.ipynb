```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
```

```python
name_i = "/home/mrinalmanu/Desktop/analyse_pubmed_search_results/FINAL/topics/neuroscience_personality/ready_NEU_PER_DF_pre_cleaned"
name_f = "/home/mrinalmanu/Desktop/analyse_pubmed_search_results/FINAL/topics/neuroscience_personality/ready_NEU_PER_DF_pre_cleaned_processed"
```

```python
# DF imports

countries_dataframe = "/home/mrinalmanu/Desktop/analyse_pubmed_search_results/FINAL/countries_2"
coordinates_dataframe = "/home/mrinalmanu/Desktop/analyse_pubmed_search_results/FINAL/countries_coods.tsv"
```

```python
# Load in the dataframe
df = pd.read_csv(name_i)
```

```python
df.head(1)
```

```python
# need to replace 3-iso to 2-iso

df['Countries'] = df['Countries'].str.replace('USA','US').str.replace(' +',' ')
df['Countries'] = df['Countries'].str.replace('ETH','ET').str.replace(' +',' ')
df['Countries'] = df['Countries'].str.replace('ECU','EC').str.replace(' +',' ')
df['Countries'] = df['Countries'].str.replace('FIN','FI').str.replace(' +',' ')


df['Countries'] = df['Countries'].str.replace('GRC','GR').str.replace(' +',' ')
df['Countries'] = df['Countries'].str.replace('BIH','BA').str.replace(' +',' ')

df['Countries'] = df['Countries'].str.replace('NRU','NR').str.replace(' +',' ')
df['Countries'] = df['Countries'].str.replace('AND','AD').str.replace(' +',' ')
df['Countries'] = df['Countries'].str.replace('TJK','TJ').str.replace(' +',' ')

```

```python
iso_info = pd.read_csv(countries_dataframe)
iso_info.head(1)
```

```python
del iso_info['timezones']
del iso_info['capital']
iso_info.head()
```

```python
# attaching latitude longitude info

coods = pd.read_csv(coordinates_dataframe, sep='\t')
coods.head()
```

```python
country_info = pd.merge(coods, iso_info, left_on='country', right_on='code')

del country_info['country']
del country_info['name_y']
country_info
```

```python
df_n = pd.merge(df, country_info, left_on='Countries', right_on='code')

del df_n['Countries']

df_n.head()
```

```python
df_n.to_csv(name_f, index=False)
```
